```python
def one_hot_encoding(column):
    col_unique=df[column].unique()
    col_temp=[]
    for x in range(len(col_unique)):
        col_unique[x]=col_unique[x].replace(' ','_')
        col_temp.append(column+'_'+col_unique[x])

    for x in range(len(col_temp)):
        df[col_temp[x]]=0

    for x in range(len(col_temp)):
        df[col_temp[x]]=(df[column]==col_unique[x]).astype(int)

    df.drop(column,inplace=True,axis=1)        
  
```

```python
one_hot_encoding('MultipleLines')
one_hot_encoding('InternetService')
one_hot_encoding('OnlineSecurity')
one_hot_encoding('OnlineBackup')
one_hot_encoding('DeviceProtection')
one_hot_encoding('TechSupport')
one_hot_encoding('StreamingTV')
one_hot_encoding('StreamingMovies')
one_hot_encoding('Contract')
one_hot_encoding('PaymentMethod')
```
