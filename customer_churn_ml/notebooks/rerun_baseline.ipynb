```python
import pandas as pd

df=pd.read_csv("/Users/shinojphilipjohn/Github/data-engg-science-projects/customer_churn_ml/dataset/Telco-Customer-Churn_feature_engg_2.csv")
df
```

```python
from sklearn.linear_model import LogisticRegression
X=df.drop(columns='Churn_Yes',axis=1)
y=df['Churn_Yes']

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=17,stratify=y)

print(X_train,X_test,y_train,y_test)

model_rerun=LogisticRegression(max_iter=1000,random_state=44)
model_rerun.fit(X_train,y_train)
y_pred=model_rerun.predict(X_test)
y_prob=model_rerun.predict_proba(X_test)[:,1]
```

```python
from sklearn.metrics import accuracy_score,roc_auc_score,precision_score,recall_score,f1_score,confusion_matrix

print("Accuracy: ", accuracy_score(y_test,y_pred))
print("Precision: ",precision_score(y_test,y_pred))
print("Recall: ",recall_score(y_test,y_pred))
print("F1 Score: ",f1_score(y_test,y_pred))
print("Confusion Matrix: ",confusion_matrix(y_test,y_pred))
print("ROC AUC: ",roc_auc_score(y_test,y_prob))
```

```python
#improve model by lowering the threshold to 0.35
threshold=0.35

y_pred_adjusted=y_prob>threshold

print("Accuracy:", accuracy_score(y_test, y_pred_adjusted))
print("Precision:", precision_score(y_test, y_pred_adjusted))
print("Recall:", recall_score(y_test, y_pred_adjusted))
print("F1 Score:", f1_score(y_test, y_pred_adjusted))
print(confusion_matrix(y_test, y_pred_adjusted))
```

```python
#improve model by lowering the threshold to 0.3
threshold=0.3


y_pred_adjusted=y_prob>threshold

print("Accuracy:", accuracy_score(y_test, y_pred_adjusted))
print("Precision:", precision_score(y_test, y_pred_adjusted))
print("Recall:", recall_score(y_test, y_pred_adjusted))
print("F1 Score:", f1_score(y_test, y_pred_adjusted))
print(confusion_matrix(y_test, y_pred_adjusted))
```

```python
#Lowering the threshold significantly improved the recall thus depending on if the business wants an agressive strategy would suggest 0.3 otherwise 0.35 would maintain a balance between
#precision and recall.
```
