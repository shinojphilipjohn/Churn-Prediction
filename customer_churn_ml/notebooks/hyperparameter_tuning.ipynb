```python
import pandas as pd

df=pd.read_csv("/Users/shinojphilipjohn/Github/data-engg-science-projects/customer_churn_ml/dataset/Telco-Customer-Churn_feature_engg_2.csv")
df
```

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

df.drop([col for col in df.columns if 'tenure_bucket' in col], axis=1, inplace=True)
X=df.drop(columns='Churn_Yes',axis=1)
y=df['Churn_Yes']

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=17,stratify=y)

print(X_train,X_test,y_train,y_test)

model_rfc=RandomForestClassifier(random_state=55,class_weight='balanced')

param_distributions={
    'n_estimators':[1000,5000,10000]
    ,'max_depth':[0,5,7,9,10,15,17,20]
    ,'min_samples_split':[5,7,14,18,20]
    ,'min_samples_leaf':[1,2,4,6,7]
    ,'max_features':['sqrt', 'log2', None]
    ,'criterion':['gini','entropy','log_loss']
    ,'bootstrap':[True,False]

}

rscv_rf=RandomizedSearchCV(
    model_rfc,
    param_distributions=param_distributions,
    n_iter=33,
    scoring='f1',
    cv=5,
    verbose=2,
    n_jobs=-1,
    random_state=33
)

rscv_rf.fit(X_train,y_train)

print('Best params found:')
print(rscv_rf.best_params_)

print('Best score:')
print(rscv_rf.best_score_)

best_rfm=rscv_rf.best_estimator_

y_pred=best_rfm.predict(X_test)
ypred_prob=best_rfm.predict_proba(X_test)[:,1]
```

```python
from sklearn.metrics import accuracy_score,roc_auc_score,precision_score,recall_score,f1_score,confusion_matrix

print("Accuracy: ", accuracy_score(y_test,y_pred))
print("Precision: ",precision_score(y_test,y_pred))
print("Recall: ",recall_score(y_test,y_pred))
print("F1 Score: ",f1_score(y_test,y_pred))
print("Confusion Matrix: ",confusion_matrix(y_test,y_pred))
print("ROC AUC: ",roc_auc_score(y_test,ypred_prob))
```
